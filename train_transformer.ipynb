Pilihan model: **RT-DETR-L** (resmi tersedia di Ultralytics) agar auto-download berhasil.
Catatan: jika Anda punya bobot RT-DETR yang lebih ringan, tinggal ganti `MODEL_PATH`.


```python
import importlib.util
import sys
import subprocess

if importlib.util.find_spec('ultralytics') is None:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-U', 'ultralytics'])
```

```python
import ultralytics
print(ultralytics.__version__)
```

```python
from ultralytics import RTDETR
import torch

MODEL_PATH = 'rtdetr-l.pt'
DATA_YAML = 'data.yaml'
IMG_SIZE = 640
EPOCHS = 100
BATCH = 4
WORKERS = 0
DEVICE = 0 if torch.cuda.is_available() else 'cpu'

model = RTDETR(MODEL_PATH)
results = model.train(
    data=DATA_YAML,
    imgsz=IMG_SIZE,
    epochs=EPOCHS,
    batch=BATCH,
    workers=WORKERS,
    device=DEVICE,
    project='runs/train',
    name='rtdetr_lele'
)
```

```python
from ultralytics import RTDETR

# 1. Load model original PyTorch (.pt) hasil training di laptop Anda
model = RTDETR('runs/detect/runs/train/rtdetr_lele-2/weights/best.pt')

# 2. Re-export ke NCNN dengan parameter penyederhanaan
model.export(
    format='ncnn',
    imgsz=640,         # Tentukan ukuran gambar statis (sesuai resolusi training Anda)
    dynamic=False,     # PENTING: Nonaktifkan shape dinamis (wajib False agar statis)
    simplify=True,     # PENTING: Mengaktifkan onnx-simplifier untuk menyederhanakan graph
    batch=1            # Kunci ukuran batch menjadi 1
)
```
