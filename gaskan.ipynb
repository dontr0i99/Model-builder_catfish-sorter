```python
import os

LABEL_DIR = "dataset_split/valid/labels"  # ganti sesuai (train/valid/test nanti semua dijalankan)

# mapping lama → baru
normal_classes = {'0', '1', '2'}   # 3,4,5
anomali_classes = {'3', '4', '5'}  # 6,7,8

for file in os.listdir(LABEL_DIR):
    if not file.endswith(".txt"):
        continue

    path = os.path.join(LABEL_DIR, file)

    new_lines = []

    with open(path, "r") as f:
        lines = f.readlines()

    for line in lines:
        parts = line.strip().split()

        if len(parts) == 0:
            continue

        cls = parts[0]

        if cls in normal_classes:
            new_cls = "0"
        elif cls in anomali_classes:
            new_cls = "1"
        else:
            continue  # skip kalau aneh

        parts[0] = new_cls
        new_lines.append(" ".join(parts))

    # overwrite file
    with open(path, "w") as f:
        f.write("\n".join(new_lines))

print("Convert label selesai!")
```

```python
from ultralytics import YOLO

# load model ringan dulu
model = YOLO("yolo11n.pt")  # bisa ganti yolo11s.pt kalau mau lebih akurat

model.train(
    data="data.yaml",
    epochs=100,
    imgsz=640,
    batch=16,       # kalau error, turunin ke 8 atau 4
    device=0,       # pakai GPU, kalau CPU pakai "cpu"
    workers=4,
    patience=20,    # early stopping
    project="runs/train",
    name="lele_v1",
)
```
